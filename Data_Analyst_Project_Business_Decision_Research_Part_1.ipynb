**Importing Data dan Inspection**

Importlah dataset dari https://storage.googleapis.com/dqlab-dataset/data_retail.csv dan kemudian inspeksilah dataset tersebut dengan

1. mencetak lima data teratas saja,
2. mencetak info dataset.

```
# Mengimpor library pandas
import pandas as pd

# Membaca dataset dari URL menggunakan fungsi read_csv dengan pemisah (delimiter) berupa titik koma (;)
df = pd.read_csv('https://storage.googleapis.com/dqlab-dataset/data_retail.csv', sep = ';')

# Menampilkan lima data pertama dari dataset
print('Lima data teratas:')
print(df.head())

# Menampilkan informasi umum mengenai dataset, seperti jumlah entri, kolom, tipe data, dan penggunaan memori
print('\nInfo dataset:')
print(df.info())
```

Berdasarkan hasil pengecekan tidak ada missing velues

**Data Cleansing**

Dua kolom yang menunjukkan terjadinya transaksi tidak bertipe datetime, maka ubahlah kedua kolom tersebut ke tipe data datetime. Kemudian cetaklah kembali 5 data teratas dari dataframe df dan juga tipe data masing-masing kolomnya.

```
# Mengonversi kolom 'First_Transaction' dari format timestamp (dalam milidetik) ke format datetime
# Dibagi 1000 karena timestamp awal dalam milidetik, sedangkan pandas mengharapkan detik
df['First_Transaction'] = pd.to_datetime(df['First_Transaction']/1000, unit='s', origin='1970-01-01')

# Mengonversi kolom 'Last_Transaction' dari format timestamp (dalam milidetik) ke format datetime
df['Last_Transaction'] = pd.to_datetime(df['Last_Transaction']/1000, unit='s', origin='1970-01-01')

# Menampilkan lima baris pertama dari dataset setelah konversi timestamp
print('Lima data teratas:')
print(df.head())

# Menampilkan informasi umum tentang dataset, termasuk tipe data setelah konversi
print('\nInfo dataset:')
print(df.info())
```

Churn Customers

Untuk menentukan churn customers sesuai definisi yang telah diberikan, carilah

1. transaksi paling terakhir kapan dilakukan
2. klasifikasikanlah mana customer yang berstatus churn dan mana yang tidak.
Setelah itu cetak lima data teratas dan informasi dataset.

```
# Pengecekan transaksaksi terakhir dalam dataset
print(max(df['Last_Transaction']))

# Klasifikasikan customer yang berstatus churn atau tidak dengan boolean
df.loc[df['Last_Transaction'] <= '2018-08-01', 'is_churn'] = True
df.loc[df['Last_Transaction'] > '2018-08-01', 'is_churn'] = False

# Menampilkan lima baris pertama dari dataset
print('Lima data teratas:')
print(df.head())

# Menampilkan informasi umum tentang dataset, termasuk tipe data setelah konversi
print('\nInfo dataset:')
print(df.info())
```

1. Transaksi paling terakhir dalam dataset terjadi pada 1 Februari 2019 pukul 23:57:57.

2. Pelanggan diklasifikasikan sebagai churn atau tidak churn berdasarkan waktu transaksi terakhir mereka:

*   Churn: pelanggan yang tidak lagi melakukan transaksi sejak atau sebelum 1 Agustus 2018.
*   Tidak churn: pelanggan yang masih aktif bertransaksi setelah 1 Agustus 2018.

Mengubah value kolom is_churn menjadi numerik menggunakan Label Encoder

```
# Import library numpy dengan alias np
import numpy as np

# Import LabelEncoder dari sklearn untuk mengubah nilai kategori menjadi angka
from sklearn.preprocessing import LabelEncoder

# Inisialisasi objek LabelEncoder
LE = LabelEncoder()

# Melakukan encoding terhadap kolom 'is_churn' agar nilai kategorinya (misalnya: True/False atau 'Yes'/'No')
# dikonversi menjadi angka (biasanya 0 dan 1), lalu hasilnya disimpan kembali ke kolom yang sama
df['is_churn'] = LE.fit_transform(df['is_churn'])

# Menampilkan daftar kelas asli (sebelum diubah ke angka)
print(LE.classes_)

# Menampilkan nilai unik dalam kolom 'is_churn' setelah diubah (misalnya: [0 1])
print(np.sort(df['is_churn'].unique()))

# Mencetak baris kosong sebagai pemisah
print('')
```

Rentang jumlah transaksi	Kategori

image.png


Setelah menambahkan kolom baru untuk kategori ini dengan nama Count_Transaction_Group, maka visualisasikanlah dengan bar chart.

```
import matplotlib.pyplot as plt

plt.clf()

# Fungsi untuk mengelompokkan jumlah transaksi ke dalam kategori
def func(row):
    if row['Count_Transaction'] == 1:
        val = '1. 1'
    elif (row['Count_Transaction'] >1 and row['Count_Transaction'] <= 3):
        val ='2. 2 - 3'
    elif (row['Count_Transaction'] >3 and row['Count_Transaction'] <= 6):
        val ='3. 4 - 6'
    elif (row['Count_Transaction'] >6 and row['Count_Transaction'] <= 10):
        val ='4. 7 - 10'
    else:
        val ='5. >10'
    return val
# Menambahkan kolom baru berdasarkan hasil pengelompokan jumlah transaksi
df['Count_Transaction_Group'] = df.apply(func, axis=1)

# Membuat grafik batang untuk distribusi pelanggan berdasarkan kelompok jumlah transaksi
df_year = df.groupby(['Count_Transaction_Group'])['Customer_ID'].count()
df_year.plot(x='Count_Transaction_Group', y='Customer_ID', kind='bar', title='Customer Distribution by Count Transaction Group')
plt.xlabel('Count_Transaction_Group')
plt.ylabel('Num_of_Customer')
plt.tight_layout()
plt.show()
```

Penjelasan Grafik:

* Kelompok “2. 2 – 3” dan “3. 4 – 6” menunjukkan penurunan jumlah pelanggan, menandakan sebagian pelanggan kembali, tetapi masih dalam jumlah kecil.

* Kelompok “4. 7 – 10” sangat kecil, menunjukkan bahwa hanya sedikit pelanggan yang tergolong “frequent buyer” dalam kisaran menengah.

* Kelompok “5. >10” meningkat kembali secara signifikan, mengindikasikan adanya segmen loyal pelanggan yang rutin bertransaksi lebih dari 10 kali.

**Distribusi kategorisasi average transaction amount**

Selanjutnya, akan melakukan visualisasi dari distribusi kategorisasi average transaction amount. Kategorisasi ini dilakukan dengan mengelompokkan rata-rata besar transaksi seperti yang diperlihatkan oleh tabel berikut:

image.png

Setelah ditambahkan kolom baru untuk kategori ini dengan nama Average_Transaction_Amount_Group, maka visualisasikanlah dengan bar chart.

```
import matplotlib.pyplot as plt

plt.clf()

# Fungsi untuk mengelompokkan rata-rata jumlah transaksi ke dalam kategori nilai transaksi
def f(row):
    if (row['Average_Transaction_Amount'] >= 100000 and row['Average_Transaction_Amount'] <=200000):
        val ='1. 100.000 - 250.000'
    elif (row['Average_Transaction_Amount'] >250000 and row['Average_Transaction_Amount'] <= 500000):
        val ='2. >250.000 - 500.000'
    elif (row['Average_Transaction_Amount'] >500000 and row['Average_Transaction_Amount'] <= 750000):
        val ='3. >500.000 - 750.000'
    elif (row['Average_Transaction_Amount'] >750000 and row['Average_Transaction_Amount'] <= 1000000):
        val ='4. >750.000 - 1.000.000'
    elif (row['Average_Transaction_Amount'] >1000000 and row['Average_Transaction_Amount'] <= 2500000):
        val ='5. >1.000.000 - 2.500.000'
    elif (row['Average_Transaction_Amount'] >2500000 and row['Average_Transaction_Amount'] <= 5000000):
        val ='6. >2.500.000 - 5.000.000'
    elif (row['Average_Transaction_Amount'] >5000000 and row['Average_Transaction_Amount'] <= 10000000):
        val ='7. >5.000.000 - 10.000.000'
    else:
        val ='8. >10.000.000'
    return val

# Menambahkan kolom baru berdasarkan fungsi kategori nilai transaksi
df['Average_Transaction_Amount_Group'] = df.apply(f, axis=1)

# Mengelompokkan data berdasarkan kategori nilai transaksi dan menghitung jumlah pelanggan di tiap grup
df_year = df.groupby(['Average_Transaction_Amount_Group'])['Customer_ID'].count()

# Membuat visualisasi bar chart distribusi pelanggan berdasarkan rata-rata nilai transaksi
df_year.plot(x='Average_Transaction_Amount_Group', y='Customer_ID',kind='bar', title='Customer Distribution by Average Transaction Amount Group')

# Menambahkan label sumbu
plt.xlabel('Average_Transaction_Amount_Group')
plt.ylabel('Num_of_Customer')
plt.tight_layout()
plt.show()
```

* Jumlah pelanggan cukup tinggi juga terdapat pada kelompok: >250.000 - 500.000 dan >500.000 - 750.000

### **Coding Test: Modelling**

**Feature Columns dan Target**

Di bagian ini, selanjutnya akan menentukan feature columns dari dataset yang dimiliki, di sini dipilih kolom Average_Transaction_Amount, Count_Transaction, dan Year_Diff. Akan tetapi, kolom terakhir belum ada. Silakan dicreate dahulu kolom Year_Diff ini dan kemudian assign dataset dengan feature columns ini sebagai variabel independent X.

Untuk target tentunya persoalan costumer dengan kondisi churn atau tidak, assign dataset untuk target ini ke dalam variabe dependent y.

```
# Membuat kolom baru Year_Diff untuk menghitung durasi (dalam tahun) antara transaksi pertama dan terakhir
df['Year_Diff'] = df['Year_Last_Transaction'] - df['Year_First_Transaction']

# Mendefinisikan daftar nama kolom yang akan digunakan sebagai fitur dalam model
feature_columns = ['Average_Transaction_Amount', 'Count_Transaction', 'Year_Diff']

# Menyimpan fitur ke dalam variabel X
X = df[feature_columns]

# Menyimpan target (label churn) ke dalam variabel y
y = df['is_churn']
```

```
# Menampilkan isi variabel X yang berisi kolom-kolom fitur:
# 'Average_Transaction_Amount', 'Count_Transaction', dan 'Year_Diff'
print('Kolom Feature', X)

# Menampilkan isi variabel y yang merupakan target atau label churn,
# yaitu True untuk pelanggan churn dan False untuk pelanggan aktif
print('\nKolom Target', y)
```

**Split X dan y ke dalam bagian training dan testing**

Setelah variabel independent X dan variabel dependent y selesai dilakukan, maka pecahlah X dan y ke dalam bagian training dan testing. Bagian testing 25% dari jumlah entri data.

**Train, predict dan evaluate**

Langkah selanjutnya akan membuat model menggunakan Linear Regression, inisialisasilah model, fit, dan kemudian evaluasi model dengan menggunakan confusion matrix.

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

# Inisiasi model logreg
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train, y_train)

# Predict model
y_pred=logreg.predict(X_test)

# Evaluasi model menggunakan confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', cnf_matrix)
```
